**Read the date from files**

```python
df = spark.read.option("multiline", "true").json("Files/bing_latest_news.json")
# df now is a Spark DataFrame containing JSON data from "Files/bing_latest_news.json".
display(df)
```

**Select Value column that contains data**

```python
df = df.select("value")

display(df)
```

**TO convert single row data into multiple rows use explode function**

```python
from pyspark.sql.types import * 
from pyspark.sql.functions import *
df_exploded =df.select(explode(df["value"]).alias("json_object"))
display(df_exploded)
```

**Make a json list**

```python
df_json_list = df_exploded.toJSON().collect()
print(df_json_list[0])
```

**Convert Json string to Json dictionary**

```python
import json

json_val = json.loads(df_json_list[0])
display(json_val)
```

```python
print(json_val["json_object"]["description"])
print(json_val["json_object"]["name"])
print(json_val["json_object"]["category"])
print(json_val["json_object"]["image"]["thumbnail"]["contentUrl"])
print(json_val["json_object"]["provider"][0]["name"])
print(json_val["json_object"]["datePublished"])
print(json_val["json_object"]["url"])
```

```python
json_dic = list(map(lambda x : json.loads(x),df_json_list))
```

**Make a list of all the values**

```python
description=[]
title =[]
category=[]
image=[]
url=[]
provider=[]
datePublished=[]
for json_val in json_dic:
 try:
    if json_val["json_object"].get("category") and json_val["json_object"].get("image",{}).get("thumbnail",{}).get("contentUrl"):
        description.append(json_val["json_object"]["description"])
        title.append(json_val["json_object"]["name"])
        category.append(json_val["json_object"]["category"])
        url.append(json_val["json_object"]["image"]["thumbnail"]["contentUrl"])
        image.append(json_val["json_object"]["url"])
        provider.append(json_val["json_object"]["provider"][0]["name"])
        datePublished.append(json_val["json_object"]["datePublished"])
 except Exception as e:
    print(f"Processing Error : {e}")
```


**Make a table using Zip function**

```python
#combine a list
data = list(zip(title,description,category,image,url,provider,datePublished))

#create a schema

schema = StructType([
    StructField("title",StringType(),True),
    StructField("description",StringType(),True),
    StructField("category",StringType(),True),
    StructField("url",StringType(),True),
    StructField("image",StringType(),True),
    StructField("provider",StringType(),True),
    StructField("datePublished",StringType(),True),

]

)
df_final=spark.createDataFrame(data,schema=schema)
```

```python
df_final = df_final.withColumn("datepublished",date_format(to_date("datePublished"),"dd-MMM-yyyy"))
```

**Creating Mangaged Delta table**

```python
# df_final.write.format("delta").mode("overwrite").saveAsTable("tbl_latest_news")
```

**Implement Incremental Load**

```python
from pyspark.sql.utils import AnalysisException
table_name = "Bing_Lakehouse.dbo.tbl_latest_news"
try:
    df_final.write.format("delta").saveAsTable(table_name)
except AnalysisException:
    df_final.createOrReplaceTempView("view_df_final")
    spark.sql(f"""
    MERGE INTO {table_name} target_table 
    USING view_df_final source_view
    on target_table.url = source_view.url
    when matched and 
    source_view.title <> target_table.title or 
    source_view.description <> target_table.description or 
    source_view.category <> target_table.category or 
    source_view.image <> target_table.image or 
    source_view.provider<>target_table.provider or 
    source_view.datepublished <> target_table.datepublished 
    then update set *  
    when not matched then insert *
    """)

```
